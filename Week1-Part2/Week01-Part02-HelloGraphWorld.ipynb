We will start by loading the required libraries:

```python
import networkx as nx
import matplotlib.pyplot as plt
```

We will first create the nodelist to have the persons as show in diagram in question

```python
# Create node list data 
nodelist = {0:'Carol',1:'Andre',2:'Fernando',3:'Diane',4:'Beverly',5:'Garth',6:'Ed',7:'Heather',8:'Ike',9:'Jane'}
```

```python
# Preview nodelist
nodelist 
```

We will next create the edge-list to have the persons as show in diagram in question

```python
# Create edge list data
edgelist = [(0,1),(0,2),(0,3),(1,2),(1,3),(1,4),(2,3),(2,5),(2,7),(3,4),(3,5),(3,6),(4,5),(4,6),(6,5),(5,7),(7,8),(8,9)]
```

```python
# Preview edgelist
edgelist
```

Now we can use the edge list and the node list to create a graph object in networkx

```python
# Create a temp empty graph
g_temp = nx.Graph()
```

```python
# Add edges to the grpah from edgelist
g_temp.add_edges_from(edgelist)
```

```python
g_temp.edges()
```

```python
# we will use the relabel_nodes method and pass the temp grpah andnodelist to it.
g=nx.relabel_nodes(g_temp,nodelist)
```

```python
g.nodes()
```

```python
g.edges()
```

---

Lets print out some summary statistics before visualizing the graph.

```python
print('# of edges: {}'.format(g.number_of_edges()))
print('# of nodes: {}'.format(g.number_of_nodes()))
```

---

We will now Visualize the graph

```python
plt.figure(figsize=(10, 8))
nx.draw(g, with_labels=1, node_size=2500, node_color='red', edge_color='blue')
plt.title('Graph Representation of Nodes and Edges', size=15)
plt.show()
```
